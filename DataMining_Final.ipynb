```python
import tensorflow as tf
tf.autograph.set_verbosity(0)
from tensorflow.keras import layers

def create_ndf_model(input_shape, num_trees=10, depth=5, num_features=15):
    inputs = tf.keras.Input(shape=input_shape)
    blocks = []
    for _ in range(num_trees):
        tree = tf.keras.Sequential()
        for _ in range(depth):
            tree.add(layers.Dense(num_features, activation='relu'))
        tree.add(layers.Dense(1, activation='sigmoid'))
        blocks.append(tree(inputs))
    output = tf.keras.layers.Average()(blocks)
    model = tf.keras.Model(inputs=inputs, outputs=output)
    return model


input_shape = (17,)
num_trees = 20
depth = 5
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
num_features = X_train.shape[1]
# Create and compile the model
model = create_ndf_model(input_shape, num_trees, depth, num_features)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=15, batch_size=16, validation_split=0.3, verbose=0)

# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy}")

from sklearn.metrics import precision_score, recall_score, f1_score

# Predict the test labels
y_pred = model.predict(X_test)
y_pred_binary = (y_pred > 0.5).astype("int32")  # Convert probabilities to binary predictions

# Calculate precision, recall, and F1-score
precision = precision_score(y_test, y_pred_binary)
recall = recall_score(y_test, y_pred_binary)
f1 = f1_score(y_test, y_pred_binary)

print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Define the decision forest model
def decision_forest(input_shape, num_trees=10, depth=5, num_features=15):
    inputs = tf.keras.Input(shape=input_shape)
    blocks = []
    for _ in range(num_trees):
        tree = tf.keras.Sequential()
        for _ in range(depth):
            tree.add(layers.Dense(num_features, activation='relu'))
        tree.add(layers.Dense(1, activation='sigmoid'))
        blocks.append(tree(inputs))
    output = tf.keras.layers.Average()(blocks)
    return tf.keras.Model(inputs=inputs, outputs=output)

# Define the deep neural decision forest model
def deep_neural_decision_forest(input_shape, num_trees=10, depth=5, num_features=15):
    inputs = tf.keras.Input(shape=input_shape)
    # Neural network component
    nn = tf.keras.Sequential([
        layers.Dense(64, activation='relu'),
        layers.Dense(32, activation='relu')
    ])
    nn_output = nn(inputs)
    # Decision forest component
    df_output = decision_forest((nn_output.shape[1],), num_trees, depth, num_features)(nn_output)
    return tf.keras.Model(inputs=inputs, outputs=df_output)

# Assuming X and y are your features and target variable
# Split data into train and test sets
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

# Create and compile the model
model = deep_neural_decision_forest(input_shape=X_train.shape[1:])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=5, batch_size=16, validation_split=0.3, verbose=1)

# Evaluate the model
y_pred = model.predict(X_test)
y_pred_binary = (y_pred > 0.5).astype("int32")
accuracy = accuracy_score(y_test, y_pred_binary)
precision = precision_score(y_test, y_pred_binary)
recall = recall_score(y_test, y_pred_binary)
f1 = f1_score(y_test, y_pred_binary)

print("Test Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
